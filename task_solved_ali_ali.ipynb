```python
# little help from here :https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html
grouped_data = data_copy.groupby('Year')['TotalPay'].agg(['mean', 'count'])
print("\n average salary and Number of employees by Year:")
print(grouped_data)
```

```python
grouped_data = data_copy.groupby('JobTitle')['TotalPay'].agg(['mean', 'count'])
print("\n average salary and number of employees by Job Title:")
print(grouped_data)
```

*simple correlation analysis*

```python
#total pay is the salary so this will be solved by find the corr within this column with other
correlation = data_copy['TotalPay'].corr(data_copy['OvertimePay'])
print("\ncorrelation between Total Pay and Overtime Pay:", correlation)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalPay', y='OvertimePay', data=data_copy)
plt.title('total pay vs overtime Pay')
plt.xlabel('tetal pay')
plt.ylabel('overtime pay')
plt.show()
```
